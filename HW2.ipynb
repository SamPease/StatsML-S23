```python
import pandas
import numpy as np
from scipy.stats import bernoulli, multinomial, expon, fit, norm
from scipy import stats

import seaborn as sns
import pandas as pd
anscombe = sns.load_dataset('anscombe')
anscombe.head()
anscombe1 = anscombe[anscombe['dataset']=='I']
anscombe2 = anscombe[anscombe['dataset']=='II']
anscombe3 = anscombe[anscombe['dataset']=='III']
anscombe4 = anscombe[anscombe['dataset']=='IV']
```

```python
sns.scatterplot(data=anscombe, x='x', y='y', hue='dataset')
```

```python
sns.regplot(data=anscombe1, x='x', y='y')
sns.lmplot(data=anscombe1, x='x', y='y',order=2)
```

```python
sns.regplot(data=anscombe2, x='x', y='y')
sns.lmplot(data=anscombe2, x='x', y='y',order=2)
```

```python
sns.regplot(data=anscombe3, x='x', y='y')
sns.lmplot(data=anscombe3, x='x', y='y',order=2)
```

```python
sns.regplot(data=anscombe4, x='x', y='y')
sns.lmplot(data=anscombe4, x='x', y='y')
```
